### 5. Instancia de la libreria de OpenAI para el analisis del texto.

```python
# Download embeddings from OpenAI
embeddings = OpenAIEmbeddings()
```

```python
docsearch = FAISS.from_texts(raw_text, embeddings)
```

```python
chain = load_qa_chain(OpenAI(), chain_type="stuff")
```

### 6. Preguntas al sistema de OPenIA sobre el documento cargado.

```python
query = "Nombres de quienes firman este documento?"
docs = docsearch.similarity_search(query)
chain.run(input_documents=docs, question=query)
```
