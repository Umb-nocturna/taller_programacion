```python
from sentiment_analysis_spanish import sentiment_analysis
```

```python
sentiment = sentiment_analysis.SentimentAnalysisSpanish()
indicador = sentiment.sentiment("Es genial, podemos hacerlo todos juntos")
if indicador <= 0.4 :
  print("Negativo")
elif indicador > 0.4 and indicador < 0.6 :
  print("Neutro")
elif indicador >= 0.6 :
  print("Positivo")
```
